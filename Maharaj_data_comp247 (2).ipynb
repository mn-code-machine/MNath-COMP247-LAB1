```python
# Pandas pd
# By Maharaj Nath
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Maharaj_fruits = {
    'apple': 'red',
    'banana': 'yellow',
    'grape': 'purple',
    'orange': 'orange'
}

# Convert dictionary to Pandas Series
Maharaj_f = pd.Series(Maharaj_fruits)

# Print 2nd and 3rd items
print("Second and third items:")
print(Maharaj_f[1:3])

# Create sub-series
Maharaj_f2 = Maharaj_f[1:3]

# Print last item of sub-series using iloc
print("Last item of sub-series:")
print(Maharaj_f2.iloc[-1])
```

```python
# Handling Time
# Rainfall amounts
Maharaj_amounts = [10, 23, 24, 30]

# Create date range starting from today
dates = pd.date_range(start=pd.Timestamp.now(), periods=4, freq='H')

# Combine list and date range into Series
Maharaj_rainfall_amounts_today = pd.Series(Maharaj_amounts, index=dates)

# Plot as bar chart
Maharaj_rainfall_amounts_today.plot(kind='bar')
plt.title("Rainfall Amounts Today")
plt.xlabel("Date and Time")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Operations on Dataframes
# Create grades dataframe
Maharaj_grades = pd.DataFrame({
    'April': [65, 45, 80, 30],
    'May': [70, 55, 60, 40],
    'June': [75, 50, 85, 35],
    'July': [80, 60, 90, 45]
}, index=['John', 'Mary', 'Ali', 'Sara'])

# Average for April
print("Average for April:", Maharaj_grades['April'].mean())

# Increase all grades by 2%
Maharaj_grades = Maharaj_grades * 1.02

# Print May grades > 50%
print("May grades > 50%:\n", Maharaj_grades[Maharaj_grades['May'] > 50])

# Group failing students (average < 50%)
failing_students = Maharaj_grades[Maharaj_grades.mean(axis=1) < 50]
print("Failing students:\n", failing_students)
```

```python
# Matrix Inverse and psuedo inverse
Maharaj_array = np.arange(16).reshape(4, 4)
print("Array:\n", Maharaj_array)

# Inverse may fail if singular, using pseudo-inverse instead
pseudo_inverse = np.linalg.pinv(Maharaj_array)
print("Pseudo-inverse:\n", pseudo_inverse)
```

```python
# Solve a system of linear equations
import numpy as np

# Coefficient matrix
A = np.array([
    [2, 4, 1],
    [3, 8, 2],
    [1, 2, 3]
])

# Constants vector
B = np.array([12, 16, 3])

# Solve the system
solution = np.linalg.solve(A, B)
print("Solution [x, y, z]:", solution)

# Verify solution using allclose
check = np.allclose(np.dot(A, solution), B)
print("Solution verified:", check)


```

```python
# Matplotlib
x = np.linspace(-4, 4, 1000)
z = np.linspace(-5, 5, 1000)
y = x**2 + z**3 + 6

plt.plot(x, y)
plt.title("Polynomial_Maharaj")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Subplots
x = np.linspace(-3, 3, 100)

plt.figure(figsize=(10, 8))

# Row 1
ax1 = plt.subplot2grid((4,4), (0,0), colspan=4)
ax1.plot(x, x**2, 'g--')
ax1.set_title("x^2")

# Row 2
ax2 = plt.subplot2grid((4,4), (1,0))
ax2.plot(x, x**3, 'y')
ax2.set_title("x^3")

ax3 = plt.subplot2grid((4,4), (1,1), colspan=3)
ax3.plot(x, x**4, 'r')
ax3.set_title("x^4")

# Row 3
ax4 = plt.subplot2grid((4,4), (2,0), colspan=2)
ax4.plot(x, x**6, 'b--')
ax4.set_title("x^6")

ax5 = plt.subplot2grid((4,4), (2,2), colspan=2)
ax5.plot(x, x, color='magenta')
ax5.set_title("x")

# Row 4
ax6 = plt.subplot2grid((4,4), (3,0), colspan=4)
ax6.plot(x, x**7, 'r:')
ax6.set_title("x^7")

plt.tight_layout()
plt.show()
```

```python
# Drawing text
x = np.linspace(-4, 4, 1000)
z = np.linspace(-5, 5, 1000)
y = x**2 + z**3 + 6

plt.plot(x, y)
plt.scatter(1, y[250])
plt.text(1, y[250], "new_point_Maharaj (1, y)")
plt.title("Polynomial_Maharaj")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
